### 3.1 What is the correlation between all factors and exam scores?

In this section, we aim to identify which factors are most strongly correlated with exam scores. To do this, we will first convert all categorical variables to numerical form, allowing us to apply a correlation function. Then, we will generate a correlation matrix to quantify the relationship between each factor and the exam score.

This analysis will give us an overview of which factors (e.g., study habits, family background, lifestyle) are most closely associated with student performance, and help guide further exploration in the following sections.


```python
#Create new dataframe to perform the following correlation
student_performance_df_for_corr =  student_performance_df.select_dtypes(exclude=["object"])
```

#### Handling Categorical Data and Mapping Challenges

We've encountered a challenge: to run a correlation function between all columns and the exam scores, we can only work with numerical data. However, a significant portion of our factors are categorical or stored as object data types, which complicates the process.

We also face additional challenges regarding the inconsistent ordering of the data:

1. Inconsistent Relationships with Exam Scores:
   - For example, in the Extracurricular column, "No" is associated with lower test scores, while "Yes" is linked to higher scores.
   - Conversely, in the Learning Disability column, "Yes" is associated with worse scores, the opposite of the extracurricular column.

2. Variation in Factor Levels:
   - The factor levels differ across columns. For example:
     - Parental Involvement uses "Low", "Medium", and "High".
     - Parental Education Level uses "High School", "College", and "Postgraduate."

While we could manually map these categorical values to numerical equivalents, as we did before, this approach becomes tedious and repetitive when dealing with many columns.

To ensure our code remains reusable and efficient, we need to explore more dynamic methods for converting categorical columns to numerical data.

```python
#Create dataframe to hold the ordered values
categorical_order_df = pd.DataFrame()
```

```python
#Create dataframe of the non numeric categorical columns
categorical_columns = student_performance_df.select_dtypes(exclude=["int", "float"]).columns
```

In this code block, we are addressing the challenge of dealing with categorical variables by creating an ordered representation of each non-numeric column based on the average exam scores of the respective categories. This allows us to identify patterns and trends among different groups in relation to exam performance.

For each categorical column, the data is grouped by unique categories in that column and sorted by the mean test score.  The `groupby(column)["Exam_Score"].mean().sort_values()` function handles this. Because the groupby function turns the unique values into indexes we convert them into the DataFrame (`category_order`). The column name is updated to match the orignal column name to ensure that it is traceable.

Finally we build a consolidated DataFrame (`categorical_order_df`) that stores the new ordered categories by each iteration of the loop.

This apporach allows us the asses the relationship between these factors for each individual column, preparing us for the steps we are going to take further on in the analysis.




```python
for column in categorical_columns:
    #Create grouped dataframe for sorted mean values
    mean_exam_score_by_category = student_performance_df.groupby(column)["Exam_Score"].mean().sort_values()
    category_order = pd.DataFrame(mean_exam_score_by_category.index)  # Turn our indexes into a DataFrame
    category_order.columns = [column]  # This will keep track of which column the index belongs to
    categorical_order_df = pd.concat([categorical_order_df, category_order], axis=1) if not categorical_order_df.empty else category_order #check if dataframe is empty for concatination
```

```python
#Check to see if our mapping works
categorical_order_df
```

We define a function that takes in the category value and a list of the ordered categories. It then returns the index of the matching value from our ordered list, giving its correct corresponding numerical representation.

This approach maintains the hierarchical relationship between categories (e.g., Low < Medium < High) when converting.

We then iterate through each column in the `categorical_order_df` and use the `apply()` method to apply our `map_category_to_index` function.

This dynamic approach allows us to efficiently handle multiple categorical columns, mapping them to numerical values while maintaining their natural order.

Although we will only be using this for our correlation heatmap, this gives us the option to perform other numerical operations and include these columns in machine-learning models if we decide to. 

The benefits of this method are ensuring the natural hierarchy of the categories is maintained as well as allowing us to reuse this function making scalable solutions for large datasets with many ordered categorical variables.


```python
# define function to map each item to corresponding index
def map_category_to_index(category_value, ordered_categories):
    # Return the index of the matching value in averagetestscores
    matching_indices  = ordered_categories[ordered_categories == category_value].index
    return matching_indices.item()
# Iterate over columns in index_df and apply the mapping function we created
for newcolumn in categorical_order_df.columns:
    student_performance_df_for_corr[newcolumn] = student_performance_df[newcolumn].apply(
        lambda category_value: map_category_to_index(category_value, categorical_order_df[newcolumn])  
    )
```

We create a correlation matrix between all of our new columns.

```python
correlation_matrix = student_performance_df_for_corr.corr()
```

We can finally create a heatmap that displays the relationship between all of our columns. In order to display their correlations with the Exam Scroe column specifically we point to that column in our heatmap. We sort the values so that we can see the hierarchy of importance of factors more easily.

```python
sns.heatmap(correlation_matrix[['Exam_Score']].sort_values(by='Exam_Score', ascending=False), annot=True)
plt.title('Correlation of ALL Factors with Exam Score')
plt.show()
```

#### Analysis: Correlation between Factors and Exam scores

Our analysis revealed a strong positive relationship between key study habits—such as hours studied and participation in tutoring sessions—and exam performance. For the majority of students, increasing study hours led to consistent improvements in exam scores, showing no clear signs of diminishing returns. Similarly, students who attended tutoring sessions achieved higher average exam scores than those who did not, highlighting the importance of structured, guided learning in driving academic success.

However, two interesting patterns emerged:

Outliers in Study Hours: While most students benefited from more study hours, those who studied excessively—typically over 30 hours—began to show a drop in the top performers. In fact, of the students that achieved top grades, only one studied over 30 hours. This suggests that after a certain threshold, students may encounter cognitive fatigue, burnout, or study inefficiencies that negatively impact their scores. 

Diminishing Returns on Tutoring Sessions: While tutoring has a clear positive impact, diminishing returns were observed after six sessions. Students who attended up to 6 tutoring sessions saw noticeable improvements in their exam scores, but those who attended more than 6 sessions did not experience the same level of improvement. This suggests that while tutoring is beneficial, there is a limit to its effectiveness, and students may reach a point where additional sessions provide little added value.

#### Conclusion

In conclusion, while most students benefit from both increased study hours and tutoring sessions, diminishing returns exist for both study hours and tutoring after certain thresholds. For students studying over 30 hours, cognitive fatigue, burnout, or ineffective study methods may hinder their ability to achieve top scores. Similarly, for students attending more than 6 tutoring sessions, the impact of additional sessions decreases, suggesting that students may need to focus on quality of learning rather than quantity.

For students to achieve optimal results, a balanced approach is likely more effective than simply increasing study time or tutoring sessions. This approach should incorporate effective study techniques, adequate rest, and stress management to prevent diminishing returns and maximize academic success. Furthermore this approach should prioritize attendance in class as this is directly correlated to an increase in scores in both the main group and in the outliers.

This highlights again that simply increasing study hours without proper support structures and more importantly regular attendance may lead to inefficient study habits and lower test scores. 
4. Family-related factors impact on exam scores:
   - Family-related factors, such as family income and parental education, had a weaker correlation with exam scores compared to study habits and attendance. While students from higher-income families performed slightly better on average, the data suggests that personal academic habits were more crucial in determining exam performance than background factors. 
   
### Implications and Recommendations

Based on our findings, it is clear that consistent attendance, effective study techniques, and balanced use of tutoring sessions are key to improving student performance. While study hours are important, attendance consistently emerged as one of the strongest predictors of academic success. Students with higher attendance generally performed better, and this factor influenced exam scores across the board, even more so than study hours alone. Students who studied more than 30 hours did not achieve higher scores without regular attendance.

We recommend that students prioritize consistent attendance as a foundational habit. Beyond that, focusing on study quality—rather than simply increasing hours studied—can help avoid the diminishing returns we observed when students studied excessively. Schools should encourage students to attend tutoring sessions regularly, but without relying too heavily on them, as excessive tutoring beyond six sessions had limited additional benefits.

### Limitations and Future Work

While this analysis provides valuable insights into the factors influencing exam scores, it is limited by the lack of detailed information on study methods and extracurricular influences. Future research could explore these aspects to develop a more comprehensive understanding of academic performance. Additionally, incorporating more advanced analytical techniques like regression analysis could allow for more precise predictions of exam outcomes based on these variables.

### Final Thoughts

In conclusion, academic success is driven by a combination of good study habits, consistent attendance, and access to structured support like tutoring. Throughout the analysis, attendance proved to be a crucial factor that correlated strongly with better exam performance, but effective study strategies and moderate use of tutoring also played key roles. While family background factors do have some impact, students exert the most control over their outcomes through their commitment to attending classes regularly and utilizing effective study methods.

Achieving a balance between study effort, attendance, and quality learning strategies is critical for maximizing academic success. By focusing on these areas, students can ensure that their time is used efficiently and that they achieve the best possible results in their academic pursuits.
